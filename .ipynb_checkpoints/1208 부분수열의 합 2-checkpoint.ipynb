```python
import sys

n, s = map(int, sys.stdin.readline().strip().split())
a = list(map(int,sys.stdin.readline().split()))

front = a[:n//2]
rear = a[n//2:]

f_size = len(front)
r_size = len(rear)

front_ = []
rear_ = []

#기본 아이디어#
#[1,2,3]에서 조합될 수 있는 경우는 binary로 표현할 수 있다
#[0,0,0] -> 공집합, [0,1,0] -> 2만 사용, [1,0,1] -> 1과 3만 사용
#pow(2,3) 만큼 부분수열을 만들 수 있으므로, 각 케이스를 binrary로 변경하여 합을 구한다

for i in range(pow(2,f_size)):
    form = [0 for i in range(len(front))] #binary를 입력하기 위한 format 작성([0, 0, 0])
    bin_i = list(map(int,bin(i)[2:])) # i = 3인 경우, [1,0,1]을 반환 // i = 1 인 경우 [1]
    form[len(form)-len(bin_i):] = bin_i #format 크기에 맞춰서 101을 입력 // i = 1인 경우 [0, 0, 1]로 입력된다
    tmp = 0
    for select, value in zip(form,front): #행렬곱을 하면 편하지만, numpy를 사용할 수 없으므로
        tmp += (select * value) #각 원소를 곱하여 더한 값을 기록한다(1 * 1 + 2 * 0 + 3 * 1) // (1 * 1 + 2 * 0 + 3 * 1 )
    front_.append(tmp)
    
for i in range(pow(2,r_size)):
    form = [0 for i in range(len(rear))]
    bin_i = list(map(int,bin(i)[2:]))
    form[len(form)-len(bin_i):] = bin_i
    print(i, form)
    tmp = 0
    for select, value in zip(form,rear):
        tmp += (select * value)
    rear_.append(tmp)

front_.sort() 
rear_.sort() #부분수열의 각 합을 오름차순으로 정렬    
    
left = 0 #front_는 앞에서부터 시작
right = len(rear_) -1 #rear_는 뒤에서부터 시작
ans = 0

#front_ 와 rear_의 조합으로 s를 만드는 경우 탐색#
while (left < len(front_)) & (right >= 0):
    check = front_[left] + rear_[right]
    left_ = front_[left]
    right_ = rear_[right]
    
    if check == s:
        ans1 = 0
        ans2 = 0
        while left_ == front_[left]:
            left += 1
            ans1 += 1
            if left >= len(front_):
                break

        while right_ == rear_[right]:
            right -= 1
            ans2 += 1
            if right < 0:
                break
        ans += (ans1 * ans2)
    
    if check < s:
        while left_ == front_[left]:
            left += 1
            if left >= len(front_):
                break
    
    if check > s :
        while right_ == rear_[right]:
            right -= 1
            if right < 0:
                break
if s == 0: #front_[left] = rear_[right] = 0인 경우 중에서 원소를 하나도 사용하지 않은 경우는
    ans -= 1 #답에서 제외한다(공집합 제외)
print(ans)
```
