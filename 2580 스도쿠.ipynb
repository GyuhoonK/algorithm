```python
import sys
input = sys.stdin.readline
sys.setrecursionlimit(10 ** 6)

sdoku = []

for i in range(9):
    sdoku.append(list(map(int, input().split())))

all_num = set([i for i in range(1, 10)])

def possible_num(i,j):
    '''
    sdoku[i][j]에 대입 가능한 숫자를 반환
    '''
    x, y = (i//3) * 3, (j//3) * 3
    row = set(sdoku[i])
    col = set([sdoku[x][j] for x in range(9)])
    zone = set([sdoku[n][m] for n in range(x, x+3) for m in range(y, y+3)])
    #위 3개 존재하지 않는 숫자만 대입 가능함#
    return all_num - row - col - zone

#sdoku에서 0이 채워진 부분을 찾기
seq = [(i,j) for i in range(9) for j in range(9) if sdoku[i][j] == 0]

def fill(t):
    if t == len(seq):
        for i in range(9):
            for j in range(9):
                print(sdoku[i][j], end = ' ')
            print('')
        sys.exit(0)
    else :
        x, y = seq[t]
        num = possible_num(x,y)
        for N in num:
            sdoku[x][y] = N
            fill(t+1)
            sdoku[x][y] = 0

fill(0)
```
