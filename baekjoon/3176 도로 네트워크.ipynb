```python
import sys
import heapq
from math import inf
input = sys.stdin.readline

n = int(input().strip())
node = [[] for i in range(n+1)]
for i in range(n-1):
    p, q, w = map(int, input().strip().split())
    node[p].append([q,w])
    node[q].append([p,w])
parent = [[]for i in range(n+1)]
parent[1].append(0)
#parent[i][j] : i번 노드의 pow(2,j) 번째 부모 노드

depth = [inf for i in range(n+1)] #depth[i] : 'i번 노드'의 깊이
depth[1] = 0 #root 노드는 '1번 노드'로 지정

len_min = [[inf for i in range(n+1)] for i in range(n+1)]
len_max = [[0 for i in range(n+1)] for i in range(n+1)]
#len_min[i][j] : i노드에서 j+1번째 parent까지의 경로에서 가장 낮은 cost를 기록
Q = []
heapq.heappush(Q, 1)
while len(Q) != 0:
    cur = heapq.heappop(Q)
    for link, weight in node[cur]:
        if depth[link] > depth[cur] :#link 노드가 아직 탐색되지 않았으면
            depth[link] = depth[cur] + 1
            parent[link].append(cur)
            len_min[link][0] = weight
            len_max[link][0] = weight
            tmp = cur
            num = 0
            while True:
                if len(parent[tmp]) > 0:
                    if (parent[tmp][0] != 0):
                        parent[link].append(parent[tmp][0])
                        len_min[link][len(parent[link])-1] = min(len_min[link][len(parent[link])-2],\
                                                          len_min[parent[link][0]][0])
                        len_max[link][len(parent[link])-1] = max(len_max[link][len(parent[link])-2],\
                                                          len_max[parent[link][0]][0])
                    tmp = parent[tmp][0]
                else : break
            heapq.heappush(Q,link)

m = int(input().strip())
for i in range(m):
    a, b = map(int, input().strip().split())
    if depth[b] > depth[a]:
        tmp = b
        b = a
        a = tmp
    a_, b_ = a, b
    
    a_num = 0
    b_num = 0
    while depth[a] != depth[b]:
        a = parent[a][0]
        a_num += 1
    while a != b:
        a = parent[a][0]
        b = parent[b][0]
        a_num += 1
        b_num += 1
    print(min(len_min[a_][a_num-1], len_min[b_][b_num-1]),\
             max(len_max[a_][a_num-1], len_max[b_][b_num-1]))
```
