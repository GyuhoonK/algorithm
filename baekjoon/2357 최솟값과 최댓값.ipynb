```python
import sys
input = sys.stdin.readline
from math import inf, log2, ceil

def make_segment(i,node):
    left_start = node[i][2]
    left_end = (node[i][2]+node[i][3])//2
    right_start = (node[i][2]+node[i][3])//2
    right_end = node[i][3]
    if (2*i < len(node))  :
        if (left_start < left_end):
            node[2*i] = [min(a[left_start:left_end]),max(a[left_start:left_end]), left_start, left_end]
            make_segment(2*i,node)
        if (left_start == left_end):
            node[2*i] = [inf,0, left_start, left_end]
    if (2*i+1 < len(node)):
        if (right_start < right_end):
            node[2*i+1] = [min(a[right_start:right_end]),max(a[right_start:right_end]), right_start, right_end]
            make_segment(2*i+1,node)
        if (right_start == right_end):
            node[2*i+1] = [inf,0, right_start, right_end]
    
def find_minnode(left,right,node_i): #a[left:right]에서 최솟값을 찾는 과정
    if left == right :
        return a[left]
    if node_i > len(node): #사용되지 않는 노드를 의미
        return inf
    min_value,max_value, start, end = node[node_i]# 최솟값, 최댓값 해당 범위를 지정
    if (start > right) | (end < left): #노드담당구간과 관계 없는 경우
        return inf
    if (left <= start) & (end <= right):#노드담당구간이 left~right에 포함되는 경우
        return min_value
    else : #이외의 경우에는 children에서 해당 구간을 다시 검색한다
        return min(find_minnode(left,right,node_i * 2), find_minnode(left, right, node_i * 2 + 1))

def find_maxnode(left,right,node_i): #a[left:right]에서 최솟값을 찾는 과정
    if left == right :
        return a[left]
    if node_i > len(node): #사용되지 않는 노드를 의미
        return 0
    min_value,max_value, start, end = node[node_i]# 최솟값, 최댓값,해당 범위를 지정
    if (start > right) | (end < left): #노드담당구간과 관계 없는 경우
        return 0
    if (left <= start) & (end <= right):#노드담당구간이 left~right에 포함되는 경우
        return max_value
    else : #이외의 경우에는 children에서 해당 구간을 다시 검색한다
        return max(find_maxnode(left,right,node_i * 2), find_maxnode(left, right, node_i * 2 + 1))
    
n, m = map(int, input().strip().split())

a = []
for i in range(n):
    tmp = int(input().strip())
    a.append(tmp)

h = ceil(log2(n)) #segment tree의 height를 계산
node = [0 for i in range((1<<h+1)+1)]
node[1] = [min(a),max(a), 0, n] # a[start:end]내에서 최솟값을 저장하고 있음

make_segment(1,node)
for i in range(m):
    p, q = map(int, input().strip().split())
    p = p-1
    print(find_minnode(p,q,1), find_maxnode(p,q,1))
```
