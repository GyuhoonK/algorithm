```python
import sys
input = sys.stdin.readline

class my_heap():
    def __init__(self):
        self.m = [0 for i in range(int(1e5) + 1)]
        self.size = 0 #현재 힙에서 저장하고 있는 숫자의 개수
        
    def push(self, num):
        self.m[self.size + 1] = num #리스트의 가장 마지막에 숫자를 넣는다
        ### 부모 노드와의 숫자 비교 시작 ###
        p_idx = (self.size + 1) // 2
        idx = self.size + 1
        while self.m[idx] > self.m[p_idx] :
            if p_idx == 0:
                break
            self.m[idx], self.m[p_idx] = self.m[p_idx], self.m[idx]
            idx = p_idx
            p_idx = idx // 2
        self.size += 1
        
    def pop(self):
        val = self.m[1] #반환하기로 결정한 값
        self.m[1] = self.m[self.size] #가장 끝에 있는 값을 앞으로
        self.m[self.size] = 0
        idx = 1 
        while max(self.m[idx * 2], self.m[idx * 2 + 1]) > self.m[idx]:
            if self.m[idx * 2] > self.m[idx * 2 + 1] :
                self.m[idx * 2], self.m[idx] = self.m[idx], self.m[idx * 2]
                idx = idx * 2
            else :
                self.m[idx * 2 + 1], self.m[idx] = self.m[idx], self.m[idx * 2 + 1]
                idx = idx * 2 + 1
            if (idx > self.size)|(idx*2 > self.size)|((idx*2+1)>self.size) :
                break
        self.size -= 1
        return val

h = my_heap()
N = int(input().rstrip())
for _ in range(N):
    cmd = input().rstrip()
    if cmd != '0':
        h.push(int(cmd))
    if cmd == '0':
        if h.size == 0:
            print(0)
        else : 
            print(h.pop())
    
```
